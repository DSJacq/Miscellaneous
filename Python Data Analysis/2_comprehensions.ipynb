```python
squares = []

for i in range(10):
    squares.append(i**2)
    
squares
```

```python
squares = [i ** 2 for i in range(10)]
squares
```

```python
squares3 = [i**2 for i in range(30) if i % 3 == 0]
squares3
```

```python
squares3_dict = {i: i**2 for i in range(30) if i % 3 == 0}
squares3_dict
```
